```python
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import matplotlib.pyplot as plt

import SRepClass as srep

# %matplotlib widget
```

```python
# Pick an example for testing
test_folder = '../refined_sreps/salt/neg/v06Correct/103430/header.xml'
```

```python
tmp = srep.SRep( test_folder )
```

```python
upPnts = tmp.upPoints[1::2,:]
dnPnts = tmp.downPoints[1::2,:]
crPnts = tmp.crestPoints[1::2,:]

fig = plt.figure()
ax  = fig.add_subplot( 1, 3, 1, projection='3d' )
ax.scatter( upPnts[:,0], upPnts[:,1], upPnts[:,2] )

ax = fig.add_subplot( 1, 3, 2, projection='3d' )
ax.scatter( dnPnts[:,0], dnPnts[:,1], dnPnts[:,2] )

ax = fig.add_subplot( 1, 3, 3, projection='3d' )
ax.scatter( crPnts[:,0], crPnts[:,1], crPnts[:,2] )
```

```python
fig = plt.figure()
ax  = fig.add_subplot( 1, 1, 1, projection='3d' )
ax.scatter( upPnts[:,0], upPnts[:,1], upPnts[:,2] )
ax.scatter( dnPnts[:,0], dnPnts[:,1], dnPnts[:,2] )
ax.scatter( crPnts[:,0], crPnts[:,1], crPnts[:,2] )
```

```python
allPnts = np.concatenate( [upPnts,dnPnts,crPnts], axis=0 )
allPnts_perm = np.random.permutation( allPnts )

fig = plt.figure()
ax  = fig.add_subplot( 1, 2, 1, projection='3d' )
ax.scatter( allPnts[:,0], allPnts[:,1], allPnts[:,2] )

ax  = fig.add_subplot( 1, 2, 2, projection='3d' )
ax.scatter( allPnts_perm[:,0], allPnts_perm[:,1], allPnts_perm[:,2] )
```

```python
fig = plt.figure( figsize=(8,6))
ax  = fig.add_subplot( 1, 2, 1, projection='3d' )
ax.scatter( allPnts[:,0], allPnts[:,1], allPnts[:,2] )

ax.tick_params( axis='both', which='major', labelsize=12 )
ax.set_xlabel( 'x', fontsize="large" )
ax.set_ylabel( 'y', fontsize="large" )
ax.set_zlabel( 'z', fontsize="large" )
ax.set_title( "Example Organ\nBoundary Points", fontsize="large" )

fig.savefig( '../images/organ.png', format='png', dpi=216 )
```
